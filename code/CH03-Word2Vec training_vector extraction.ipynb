```python
# STEP 1: 데이터 로딩 및 전처리
from gensim.models import Word2Vec
```

```python
from gensim.utils import simple_preprocess
import numpy as np
from google.colab import files
```

```python
from gensim.utils import simple_preprocess
from google.colab import files

# BNC나 기타 코퍼스를 txt 형식으로 업로드
uploaded = files.upload()  # 예: bnc_corpus.txt

corpus_file = list(uploaded.keys())[0]

# Assuming the uploaded file is a .txt file
with open(corpus_file, 'r', encoding='utf-8') as f:
    lines = f.readlines()

sentences = [simple_preprocess(line) for line in lines if line.strip()]
```

```python
# STEP 2: Word2Vec 학습 (Skip-gram 방식)
model = Word2Vec(
    sentences,
    vector_size=300,
    window=2,
    min_count=10,
    sg=1,  # skip-gram
    workers=4
)
```

```python
# STEP 3: 전체 모델 저장 (모든 단어 + 학습 파라미터 포함)
model.save("word2vec_full.model")
files.download("word2vec_full.model")
```

```python
# COS 동사 목록 파일 업로드
uploaded = files.upload()  # 예: cos_verbs_137.txt

# Get the actual filename from the uploaded dictionary
corpus_file = list(uploaded.keys())[0]

# 파일 읽기 - use corpus_file instead of hardcoded filename
with open(corpus_file, "r") as f:
    cos_verbs = [line.strip().lower() for line in f if line.strip()]

# Word2Vec 모델에 존재하는 단어만 필터링
filtered_words = [w for w in cos_verbs if w in model.wv]
filtered_vectors = np.array([model.wv[w] for w in filtered_words])
```

```python
# 저장: 텍스트 (.vec)
with open("cos_filtered.vec", "w", encoding="utf-8") as f:
    f.write(f"{len(filtered_words)} {model.vector_size}\n")
    for w, vec in zip(filtered_words, filtered_vectors):
        vec_str = " ".join(map(str, vec))
        f.write(f"{w} {vec_str}\n")
files.download("cos_filtered.vec")
```

```python
# 저장: 벡터 행렬 (.npy)
np.save("cos_filtered.npy", filtered_vectors)
files.download("cos_filtered.npy")
```

```python
# 저장: KeyedVectors 객체 (.model)
model.wv.save("cos_filtered.model")
files.download("cos_filtered.model")
```
